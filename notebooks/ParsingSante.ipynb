```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By 
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC 
from selenium.webdriver.common.action_chains import ActionChains
```

```python
def get_soup(url):
    
    # Прописываем нестандартный заголовок, чтобы сайт не принял нас за бота, который парсит его данные :)
    headers = requests.utils.default_headers()
    headers.update({
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0',
    })
    
    response = requests.get(url=url, headers=headers)

    if not response.ok:
        raise ValueError("no content to parse")

    return BeautifulSoup(response.content.decode(), "lxml")
```

```python
source = "https://sante.by/"
analisys_comment = ""
#забираем ссылки на подразделы анализов hrefs

URL = "https://sante.by/laboratoriya#services"
hrefs = []
soup = get_soup(url=URL)
tags = soup.find(name="div", class_ = "StyledContent__Cards-sc-1xyd11-11 dDmDlY")

for tag in tags.find_all("div"):
    
    for a in tag.find_all("a"):

        url_capter = a.get("href")
        name_capter = a.text
        hrefs = np.append(hrefs, url_capter)
        #print(f"{url_capter = }, {name_capter = }")

#print(hrefs)
```

```python
def get_source_html():

    # Задаем заголовки правила хорошего тона и что бы быть похожим на настоящий браузер
    headers = {
        "accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9",
        "user-agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.67 Safari/537.36'
    }

    # Задаем опции для браузера что бы быть похожим на настоящий браузер
    options = webdriver.ChromeOptions()
    options.add_argument("start-maximized")
    # Отобразить окно браузера или в фоне выполнить
    # options.headless = True
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.67 Safari/537.36")
    # не закрывать браузер
    options.add_experimental_option("detach", True)
    options.add_argument("--disable-blink-features=AutomationControlled")

    browser = webdriver.Chrome(options=options)
    # задаем возможность нажимать на кнопки через обращение к actions
    actions = ActionChains(browser)

    #url = 'https://sante.by/laboratoriya'
   
    try:
        np_analisys= np.array([])
        source = "https://sante.by/"
        analisys_comment = " "
        for url_chapter in hrefs:
            url = source + url_chapter
            
            # переходим по ссылке
            browser.get(url)

            # Запускаем цикл (Пока кнопка Показать больше на экране нажимаем на нее 
            while True:
                try:
                    # Ждем до 5 секунд пока прогрузится страници или пока на ней не покажется кнопка 'Показать больше' 
                    WebDriverWait(browser, 5).until(EC.visibility_of_element_located((By.XPATH,
                     "//button[contains(text(), 'Показать больше')]")))
                    # Если кнопка 'Показать больше' появилась записываем ее в переменную find_button
                    find_button = browser.find_element(By.XPATH,
                                    "//button[contains(text(), 'Показать больше')]")
                    # Кликаем по кнопке
                    find_button.click()
                    print(find_button)
                except:
                    # если кнопки 'Показать больше' нет на странице выходим из цикла while
                    break

            # После того как понажимали на кнопку начинаем парсить страницу
            # находим корневой тег списка\таблицы dl, т.е получаем всю таблицу    
            section_table = browser.find_element(By.ID, "pricelist")
            # находим все имена услуг
            names = section_table.find_elements(By.TAG_NAME, "dt")
            # находим все стоимости услуг
            prices = section_table.find_elements(By.TAG_NAME, "dd")
            #my_dict = dict()

            for i, name in enumerate(names):
               # находим все стоимости услуг
                analisys_name = name.text # к этому значению можно применить regex
                analisys_cost = prices[i].text # к этому значению нужно применить regex и перевести его в число
               #my_dict[final_name] =  final_price #формируем словарь
                np_analisys = np.append(np_analisys, (source, url_chapter, analisys_name, analisys_cost, analisys_comment))
            # возвращаем словарь такого вида key = 'ПАКЕТ "Первичное обследование щитовидной железы" от' value = '29.00 рублей'
        return np_analisys.reshape(-1, 5)
    except Exception as ex:
        print("ex")
    finally:
        browser.close()
        browser.quit()

def main():
    result = get_source_html()
   
    print(result)

if __name__ == '__main__':
    main()
```
